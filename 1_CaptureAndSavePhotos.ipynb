```python
import cv2,numpy as np
```

```python

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)#1 is external webcam


id = int(input('Enter user id:'))
i=0 #to save all the clicked images

while(True):
    ret, frame = cap.read()
    frame = np.array(frame, dtype=np.uint8)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5, 0)
    for (x,y,w,h) in faces:
            cv2.rectangle(gray, (x,y), (x+w, y+h), (255,0,0), 2)
    
    
    cv2.imshow("imshow",gray)
    key=cv2.waitKey(1)
    if key==ord('q'):#'q' to quit
        break
    if key==ord('c'):# 'c' to capture 1 frame at a time
        i=i+1
        cv2.putText(gray[y:y+h,x:x+w], str(i), (10, 30), cv2.FONT_HERSHEY_PLAIN,1.5 , (0,0,0),2)
        
        cv2.imshow("imshow2",gray)# frames which are saved in folder 
        cv2.imwrite('FaceRecognizer\\Images\\User.'+str(id)+ "." +str(i)+ ".jpg", gray[y:y+h,x:x+w])
        print("Wrote Image")
        
cap.release()
cv2.destroyAllWindows()
```
